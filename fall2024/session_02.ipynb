# Step 2 - Explore student age

```python
yob_l = list(df["8. My Birth Year"])
yob_l
```

```python
cleansed_yob_l = []  # Empty list
cleansed_yob_l       # the same as print(cleansed_yob)
```

```python
for yob in yob_l:
    cleansed_yob = yob[-4:]
    cleansed_yob_l.append(cleansed_yob)

cleansed_yob_l
```

```python
age_list = []

for yob in cleansed_yob_l:
    age = 2024 - int(yob)
    age_list.append(age)

age_list
```

```python
min(age_list)
```

```python
max(age_list)
```

```python
total_age = 0

for age in age_list:
    total_age = total_age + age

total_age
```

```python
average = total_age / len(age_list)

round(average)
```

```python
total_age = 0

for age in age_list:
    if age >= 35 or age <= 22:
        print("Age", age, "is skipped")
    else:
       total_age = total_age + age

print("Total age is", total_age)

average = round(total_age / len(age_list))

print("Average is", average)
```

```python
total_age = 0

for age in age_list:
    if age >= 35 or age <= 22:
        print("Age", age, "is skipped")
#        continue

    total_age = total_age + age   # the age out of range was not skipped without the "continue"

print("Total age is", total_age)

average = round(total_age / len(age_list))

print("Average is", average)
```
